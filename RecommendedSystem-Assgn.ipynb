```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("books.csv", encoding=('ISO-8859-1'))
df.head()
```

```python
df1=df.drop('Unnamed: 0', axis=1)
df1.head()
```

```python
data=df1.rename({'User.ID':'Userid','Book.Title':'Booktitle','Book.Rating':'Bookrating'}, axis=1)
data.head()
```

```python
data.info()
```

```python
len(data.Userid.unique())
```

```python
len(data.Booktitle.unique())
```

```python
data1=data.pivot_table(index='Userid',columns='Booktitle',values='Bookrating').reset_index(drop=True)
data1
```

```python
data1.index=data.Userid.unique()
data1
```

```python
data1.fillna(0, inplace=True)
```

```python
data1
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data2= 1 - pairwise_distances(data1.values, metric ='cosine')
```

```python
data2
```

```python
data3=pd.DataFrame(data2)
```

```python
data3.index=data.Userid.unique()
data3.columns=data.Userid.unique()
```

```python
data3.iloc[0:10,0:10]
```

```python
np.fill_diagonal(data2,0)
data3.iloc[0:10,0:10]
```
